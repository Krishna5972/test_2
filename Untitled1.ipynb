```python
def changes(df,col):
    df['ema_signal']=df.apply(lambda x: 1 if x['entry'] > x[col] else -1,axis=1)
    df['pos_signal']=df.apply(lambda x:1 if x['signal']=='Buy' and x['ema_signal']==1 else (1 if x['signal']=='Sell' and x['ema_signal']==-1 else 0),axis=1)
    df_side=df[df['pos_signal']==1]
    return df_side
```
